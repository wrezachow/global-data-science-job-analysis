```python
# Cell 1 — Imports
import pandas as pd
import ast
```

### Loading data to dataframe

```python
df = pd.read_csv("data/scraped_jobs_raw.csv")
df.head(20)
```

```python
# 4290 rows and 12 Columns
# Skills shown as str even thoough it has to be list
# Salary is a number

print(df.shape)
print(df.dtypes)
```

### Convert 'Skills' to List

```python
df["skills"] = df["skills"].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])
```

```python
df.head()
```

```python
df.tail()
```

### Filling up NaN with a default date


```python
df["date_posted"] = df["date_posted"].fillna("2026-03-09")
df["date_posted"] = pd.to_datetime(df["date_posted"], format="mixed", utc=True)
```

```python
df.head()
```

### Replacing "location — some remoteok rows show "🔥 Upgrade to Premium to see salary" in the location field. That's a scraper issue — remoteok was putting paywalled content into the location. Not worth re-scraping, just clean those out." with np.nan


```python
df.loc[df["location"].str.contains("Upgrade to Premium", na=False), "location"] = None
```

```python
df.tail()
```

### filling 'work_mode'

```python
df.loc[df["location"].str.contains("worldwide|anywhere|remote", case=False, na=False), "work_mode"] = "remote"
```

```python
df.loc[~df["location"].str.contains("worldwide|anywhere|remote", case=False, na=False), "work_mode"] = "onsite"
```

```python
df.head()
```

```python
df["work_mode"].value_counts()
```

### Salary field cleanup

```python
df["salary_min"] = df["salary"].str.extract(r'\$(\d+[\d,]*)').replace(',', '', regex=True).astype(float)
df["salary_max"] = df["salary"].str.extract(r'-\s*\$(\d+[\d,]*)').replace(',', '', regex=True).astype(float)

df.rename(columns={"salary": "salary_avg"}, inplace=True)
df["salary_avg"] = (df["salary_min"] + df["salary_max"]) / 2
```

```python
df.head()
```

### Dropping duplicates


```python
print(f"Before: {df.shape}")
df.drop_duplicates(subset=['job_url'], inplace=True)
print(f"After: {df.shape}")
```

### Saving clean file


```python
df.to_csv("data/clean_jobs.csv", index=False)
```
